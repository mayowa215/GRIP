```python
sns.countplot(x = 'Ship Mode', data = superstore, palette = "Set1")
plt.title('Preferred Ship Mode')
plt.show()
```

```python
ship_mode_group = superstore.groupby("Ship Mode").sum()
ship_mode_group.reset_index(inplace = True)
print(ship_mode_group)
```

```python
ship_mode_group.plot(x="Ship Mode", y=['Sales','Profit'], kind = 'bar')
plt.title('Sales Vs Profit on Ship Mode')
plt.ylabel('Count')
```

Observation:

1. Standard Class Ship Mode has the highest sold items and profit.
2. Same Day Ship Mode has the lowest sold items and profit.

## Insights

### Weak Areas

1. In Furniture category, Furnishings and Bookcases yield low Profit. Bookcases and Tables yield no profit but loss.
2. For Office Supplies category, Fasteners and Labels have low sales. Fasteners have low Profit and Supplies have no Profit but loss
3. For Technology category, Machines have low profit from sales.
4. Southern Region yeilds low Sales and Central Region yields low Profit.

## Business Problems

1. Furnishings with high discount yield low Sales.
2. Discount on Office Supplies is high but yields low Sales. 

## Recommendation

1. Business strategy on Technology category should be preserved because with low discount, the category yields high Sales.
2. Discounts on Office Supplies category should be removed or investigated.
3. Business Policies on Southern and central Regions should be investigated.
